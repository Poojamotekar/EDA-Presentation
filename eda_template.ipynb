<div class="alert alert-block alert-info">
Tip # 3

How to interpret basic statistics</div>

+ Measures of central tendency - Mean, median, mode
+ Measures of spread / dispersion - SD, var, range, quartiles, percentiles
+ Meaures of frequency - Frequency

+ **Range:** defines the difference between the highest and lowest values.
+ **Variance**: measures how far each number in the set is from the mean and thus from every other number in the set.
+ **Standard deviation:** The standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance
+ **Quartiles:** A quartile is a statistical term that describes a division of observations into four defined intervals based on the values of the data and how they compare to the entire set of observations.
+ **Percentiles:** same but divided in 100 groups.

### Exercise 1: What is the [Inter Quartile Range (IQR)](https://medium.com/@vinitasilaparasetty/quartiles-for-beginners-in-data-science-2ca5a640b07b)? What conclusions can we draw from `describe`?

Why is the IQR useful?

Robustness to Outliers: Unlike other measures of spread (such as the range or standard deviation), the IQR is not affected by extreme outliers. It focuses on the values at the 25th and 75th percentiles, making it resistant to extreme data points.
Comparison Between Datasets: The IQR allows us to compare the spread of values between different datasets. Larger IQRs indicate greater variability, while smaller IQRs suggest tighter clustering of data points.
For example, consider a dataset of plant heights (in inches):

[ {Dataset: } 1, 4, 8, 11, 13, 17, 19, 19, 20, 23, 24, 24, 25, 28, 29, 31, 32 ]

The IQR for this dataset is approximately 14.5 inches. This means that the middle 50% of plant heights falls within a range of 14.5 inches.

### Exercise 2: How do we interpret these histograms?

Shape of the Distribution
Skewness and kurtosis provide insights into the shape of the data distribution. Skewness indicates the asymmetry, and kurtosis tells about the "tailedness" or how peaked the distribution is.

```python
# your solution
num.overall.skew().round(2)
```

- Skewness = 0: The distribution is perfectly symmetrical, resembling a normal distribution.
- Skewness > 0: The distribution is positively skewed (or right-skewed), meaning there is a longer tail on the right side of the distribution. The mass of the distribution is concentrated on the left of the figure.
- Skewness < 0: The distribution is negatively skewed (or left-skewed), meaning there is a longer tail on the left side of the distribution. The mass of the distribution is concentrated on the right of the figure.

```python
num.overall.kurtosis().round(2)
```

Excess Kurtosis = 0 (or Kurtosis = 3): The distribution has the same kurtosis as a normal distribution, known as mesokurtic. It indicates a distribution with tails similar to the normal distribution.
Excess Kurtosis > 0 (or Kurtosis > 3): The distribution is leptokurtic. It has fatter tails and a sharper peak than the normal distribution, indicating more extreme values (outliers) than expected in a normal distribution.
Excess Kurtosis < 0 (or Kurtosis < 3): The distribution is platykurtic. It has thinner tails and a flatter peak than the normal distribution, indicating fewer extreme values than expected in a normal distribution.

### Exercise 4: How to interpret this correlation matrix? Which other correlation methods we know?

Correlation Coefficient:
A value of 1 indicates a perfect positive correlation, meaning that as one variable increases, the other variable also increases proportionally.
A value of -1 indicates a perfect negative correlation, meaning that as one variable increases, the other variable decreases proportionally.
A value of 0 indicates no correlation between the variables.

Interpretation:
Form above , the correlation coefficient between "potential" and "value(in_euro)" is 0.53, indicating a moderately positive correlation. This suggests that as a player's potential increases, their market value tends to increase as well.
Similarly, the correlation coefficient between "age" and "weight(in_kg)" is 0.22, indicating a positive correlation. This suggests that older players tend to have higher weights.
The correlation coefficient between "height(in_cm)" and "weight(in_kg)" is 0.75, indicating a strong positive correlation. This suggests that taller players tend to weigh more.
Conversely, the correlation coefficient between "height(in_cm)" and "weak_foot_rating" is -0.16, indicating a weak negative correlation. This suggests that there is a slight tendency for taller players to have slightly lower weak foot ratings.
